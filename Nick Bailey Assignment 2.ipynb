Recall that in PHYS 207, you were often asked to plot displacement as a function of time. For example, for a baseball thrown directly upward you would plot $y(t)$, and for a block on a horizontal spring you would plot $x(t)$. Here we will examine how the displacement and time axes "scissor together" at relativistic speeds.

Imagine that frame $K^{\prime}$ is moving in the $+x$ direction at speed $v$ relative to frame $K$. You are at rest in frame $K$. Let's figure out how the set of $(t^{\prime},x^{\prime})$ axes would look to you when the frames' origins $O$ and $O^{\prime}$ coincide.

<ol>
    <li>Use the fact that $x^{\prime} = 0$ on the $t^{\prime}$ axis to define a line $x(t)$ in frame $K$ that follows the $t^{\prime}$ axis. Write a function that computes points on the line. $v$ will be one of your function inputs.</li>
</ol>

```python
# Writing the function:
def x_of_t(t_prime, v, c):
    beta = v/c
    gamma = 1/np.sqrt(1-(beta**2)) # Defining gamma from the Lorentz transformation
    x1 = gamma*(v*t_prime) # Calculating x(t)
    
    return x1

# Using the function
tprime_values = np.linspace(0, 10, 100)  # time values
v = float(input("Input a velocity relative to c (0<v<1): "))*c  # Relative velocity (relative to the speed of light)
c = 3e8 # Maximum theoretical velocity (v is in terms of c so set c=1 for simplicity)

# Calculate corresponding x values for each time
x_values = x_of_t(tprime_values, v, c)
```

Use your functions from problems 1 and 2 to overplot the $t^{\prime}$ and $x^{\prime}$ axes <strong>as a function of $ct$</strong> for $v = 0.2c$.
```python
# Function to compute points on the line x(t) for the t' axis
def x_of_t(t, v, c):
    beta = v/c
    gamma = 1/np.sqrt(1-(beta**2)) # Defining gamma from the Lorentz transformation
    x1 = gamma*(v*t) # Calculating x(t)
    
    return x1

# Function to compute points on the line x(t) for the x' axis
def x_of_t_2(xprime, v, c):
    # Calculate x(t) for the x' axis using Lorentz transformation
    beta = v/c
    gamma = 1/np.sqrt(1-(beta**2))
    x2 = gamma*xprime
    
    return x2

# Relative velocity and speed of light
v = 0.2*c
c = 3e8  # Speed of light (setting it to 1 for simplicity)

# Set up an array of independent variables t that spans -100 to 100 seconds
t = np.linspace(0, 200, num=201, endpoint=True) - 100
# Setting up array of independent x values that follows the same range as t
#x = np.linspace(0, 200, num=201, endpoint=True) - 100
ct = c*t
xprime = ct


# Make a figure
fig, ax = plt.subplots(figsize=(7, 7))

# Plot the t' and x' axes
t_prime_values = x_of_t(t, v, c)
x_prime_values = x_of_t_2(xprime, v, c)

ax.plot(ct, t_prime_values, label="t' axis", color="red")
ax.plot(xprime, x_prime_values, label="x' axis", color="red")

# Plot points on the t' and x' axes
ax.scatter(0, 0, color="red")  # Origin point

# Labeling the t' and x' axes in red
ax.text(5e9, 8e9, "t' axis", color="red")
ax.text(8.5e9, 2.5e9, "x' axis", color="red")

# Move the left and bottom "spines" (borders of the plot) to the center
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# Take out the top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Make the two axes have the same scale
ax.set_aspect('equal')

# Set the axes ranges
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

# Show legend
ax.legend()

# Show the plot
plt.show()
```
